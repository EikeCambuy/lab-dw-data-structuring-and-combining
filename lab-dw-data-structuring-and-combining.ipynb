```python
import pandas as pd 

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"

data_1 = pd.read_csv(url)
```

```python
import pandas as pd 

url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"

data_2 = pd.read_csv(url1)
```

```python
import pandas as pd 

url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

data_3 = pd.read_csv(url2)
```

```python
data_4 = pd.concat([data_1,data_2,data_3],axis=0 , ignore_index = True)
data_4
```

```python
data_4.info()
```

```python
data_4.rename(columns={data_4.columns[1]: 'State'}, inplace=True)
data_4.info()
```

```python
data_4.columns = data_4.columns.str.capitalize()
data_4.info()
```

```python
data_4.rename(columns=lambda x : x.strip(),inplace=True)
```

```python
data_4.isnull().sum()
```

```python
data_4 = data_4.dropna(how= "all")
```

```python
data_4.columns = ["Customer","State 1","Gender 1","Education","Customer lifetime value","Income","Monthly premium auto","Number of open complaints","Policy type","Vehicle class","Total claim amount","State","Gender"]
```

```python
data_4 = data_4.drop(columns= ["State 1","Gender 1"])
```

```python
data_4.isnull().sum()
```

```python
null_data_4 = ["State","Gender"]
data_4.dropna(subset= null_data_4 , inplace=True)
```
